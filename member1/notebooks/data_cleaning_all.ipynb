```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql

# 数据库连接配置
# 替换为您的实际数据库连接信息
db_connection_str = 'mysql+pymysql://datax:DxPwd123!@BPeril:3306/DIDatax'
db_connection = create_engine(db_connection_str)

# 待检查的表列表
tables = [
    'ga_session',
    'ga_custom_dimension',
    'ga_hit',
    'ga_total',
    'ga_geo_network',
    'ga_traffic_source',
    'ga_device'
]

print("数据库连接成功。准备检查以下表：", tables)
```

```python
from IPython.display import display

def check_data_quality(table_name, engine):
    print(f"\n{'='*20} {table_name} {'='*20}")
    try:
        # 读取数据 (如果数据量非常大，建议添加 limit 或分块读取，这里假设数据量在内存可承受范围内)
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql(query, engine)
        
        # 1. 基本信息
        print(f"数据量 (Rows, Columns): {df.shape}")
        
        # 2. 重复值检查
        duplicates = df.duplicated().sum()
        print(f"完全重复行数: {duplicates}")
        
        # 3. 缺失值检查 (只显示有缺失值的列)
        missing = df.isnull().sum()
        missing_cols = missing[missing > 0]
        if not missing_cols.empty:
            print("\n缺失值统计 (列名: 缺失数):")
            print(missing_cols)
        else:
            print("\n无缺失值。")
            
        # 4. 数据预览
        print("\n前 5 行预览:")
        return df.head()
    except Exception as e:
        print(f"读取或处理表 {table_name} 时发生错误: {e}")
        return None

# 循环执行检查
for table in tables:
    display_df = check_data_quality(table, db_connection)
    if display_df is not None:
        display(display_df)
```

```python
import os

# 创建保存目录 (相对路径，相对于 notebooks 文件夹)
# 也就是 e:\Grade3\2-DataIntegration\Assignments-DI\As2\data\cleaned
output_dir = os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'cleaned'))

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"已创建目录: {output_dir}")
else:
    print(f"输出目录已存在: {output_dir}")

def clean_and_save(table_name, engine):
    print(f"正在处理表: {table_name}...")
    try:
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql(query, engine)
        
        original_count = len(df)
        
        # 1. 初步清洗：去重
        df.drop_duplicates(inplace=True)
        dedup_count = len(df)
        
        if original_count != dedup_count:
            print(f"  - 发现并删除了 {original_count - dedup_count} 行重复数据")
        else:
            print("  - 无完全重复行")

        # 2. 导出为 CSV (提供给其他组员分析使用)
        # 使用 utf-8-sig 编码以防止 Excel 打开中文乱码
        file_path = os.path.join(output_dir, f"{table_name}.csv")
        df.to_csv(file_path, index=False, encoding='utf-8-sig')
        print(f"  - 已导出至: {file_path}")
        
    except Exception as e:
        print(f"  - 处理失败: {e}")

print("-" * 30)
print("开始执行清洗与导出任务（供 Team Member 2-5 使用）...")

# 循环执行清洗并保存
for table in tables:
    clean_and_save(table, db_connection)

print("\n所有清洗后的数据文件生成完毕！")
```
